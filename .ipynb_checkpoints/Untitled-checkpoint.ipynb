# tsne clustering

```python
import torch
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import torchvision

# Load the dataset
data = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=torchvision.transforms.ToTensor())
images, labels = zip(*data)

```

```python

# Create a PyTorch DataLoader for the dataset
batch_size = 100
data_loader = torch.utils.data.DataLoader(data, batch_size=batch_size)

# Create a feature extractor using a pre-trained ResNet50 model
model = torchvision.models.resnet50(pretrained=True)
model = torch.nn.Sequential(*(list(model.children())[:-1]))
model.eval()
```

```python


# Extract features for each image using the feature extractor
features = []
with torch.no_grad():
    for i, (images, labels) in enumerate(data_loader):
        print(f'Extracting features: batch {i+1}/{len(data_loader)}')
        features_batch = model(images).squeeze()
        features.append(features_batch)
features = torch.cat(features, dim=0).numpy()

# Perform t-SNE clustering on the features
tsne = TSNE(n_components=2, random_state=0)
embedding = tsne.fit_transform(features)

# Plot the t-SNE embedding
plt.scatter(embedding[:, 0], embedding[:, 1], c=labels)
plt.show()
```
